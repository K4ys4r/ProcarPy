## Si Diamond
For this calculation all input files can be found in [test/Si_diamond file](https://github.com/K4ys4r/ProcarPy).<br>
The Band Structure has bee calculated along the high symmetry $k-$points path: $L-\Gamma-X-K-\Gamma$.
### LORBIT = 10

```python
# importing ProcarPy module
import ProcarPy
```

```python
# calling PROCARBandStructure class
Si_Bands = ProcarPy.PROCARBandStructure(filename="./LORBIT=10/PROCAR", ef=6.11994092, path=["L","G","X","K","G"], pathstyle="continuous")
```

```python
# Plotting the Total Band Structure
ProcarPy.p.rc("font",size=20)
Si_Bands.Init_Fig()
Si_Bands.getTotalBandsPlot()
Si_Bands.PlotShow()
```

```python
Si_Bands.Init_Fig()
Si_Bands.getTotalBandsPlot(lw=2,label="Si Total Bands")
Si_Bands.PlotShow(ymin=-3,ymax=3)
```

```python
# getting the band gap. 
# NELECT=8. Hence Valence Band Minimum index = 4 Conduction Band Maximum index = 5. 
Si_Bands.getBandsGap(band1=4,band2=5)
```

```python
# Plotting the. Projected Band of s and p orbitals of Si atom. 
Si_Bands.Init_Fig()
Si_Bands.getOrbitalBandsPlot(orbital="s",atom=1,scale=500,alpha=0.5,label="Si($s$)")
Si_Bands.getOrbitalBandsPlot(orbital="p",atom=1,scale=500,alpha=0.5,color="r",label="Si($p$)")
Si_Bands.PlotShow(-3,4)
```

### LORBIT = 11

```python
Si_Bands = ProcarPy.PROCARBandStructure(filename="./LORBIT=11/PROCAR", ef=6.11994092, path=["L","G","X","K","G"], pathstyle="continuous")
```

```python
# Plotting the. Projected Band of s, px, py ad pz orbitals of Si atom. 
s = 1000
Si_Bands.Init_Fig(10,5)
Si_Bands.getOrbitalBandsPlot(orbital="s",atom=1,scale=s,alpha=0.5,label="Si($s$)")
Si_Bands.getOrbitalBandsPlot(orbital="px",atom=1,scale=s,alpha=0.5,color="r",label="Si($p_x$)")
Si_Bands.PlotShow(-3,4)

Si_Bands.Init_Fig(10,5)
Si_Bands.getOrbitalBandsPlot(orbital="s",atom=1,scale=s,alpha=0.5,label="Si($s$)")
Si_Bands.getOrbitalBandsPlot(orbital="py",atom=1,scale=s,alpha=0.5,color="g",label="Si($p_y$)")
Si_Bands.PlotShow(-3,4)

Si_Bands.Init_Fig(10,5)
Si_Bands.getOrbitalBandsPlot(orbital="s",atom=1,scale=s,alpha=0.5,label="Si($s$)")
Si_Bands.getOrbitalBandsPlot(orbital="pz",atom=1,scale=s,alpha=0.5,color="k",label="Si($p_z$)")
Si_Bands.PlotShow(-3,4)

Si_Bands.Init_Fig(10,5)
Si_Bands.getOrbitalBandsPlot(orbital="px",atom=1,scale=s,alpha=0.5,color="r",label="Si($p_x$)")
Si_Bands.getOrbitalBandsPlot(orbital="py",atom=1,scale=s,alpha=0.5,color="g",label="Si($p_y$)")
Si_Bands.getOrbitalBandsPlot(orbital="pz",atom=1,scale=s,alpha=0.5,color="k",label="Si($p_z$)")
Si_Bands.PlotShow(-3,4)
```
