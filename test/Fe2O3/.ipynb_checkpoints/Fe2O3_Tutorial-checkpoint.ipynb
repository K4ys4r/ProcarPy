## Hematite ($\alpha-Fe_2O3$) Spin-Polarized calculation
For this calculation $\alpha-Fe_2O_3$ is treated as antiferromagnetic(AFM), with alternating of the spin-up and spin-down of aligned iron atoms (Fig.1).

<figure class="image">
  <img src="fe2O3.png" alt="Fe2O3 Structure" width="250">
  <center> Fig.1: The AFM Structure of $\alpha-Fe_2O_3$.</center>
</figure>

For this calculation all input files can be found in [test/Fe2O3](https://github.com/K4ys4r/ProcarPy/tree/master/test/Fe2O3).<br>
The Band Structure has bee calculated along the high symmetry $k-$points path: $H_0-L-\Gamma-S_0$.

```python
from ProcarPy import PROCARBandStructure as PB
```

```python
Fe2O3_Bands = PB(filename="./LORBIT=10/PROCAR",
               ef=3.21537059, spin=True, 
               path=["$H_0$","$L$","$\Gamma$","$S_0$"], 
               pathstyle="continuous")
```

```python
Fe2O3_Bands.Init_Fig()
Fe2O3_Bands.getTotalBandsPlot(bandspin="up",color="r",lw=2,alpha=0.5)
Fe2O3_Bands.getTotalBandsPlot(bandspin="down",color="b",lw=2,alpha=0.5)
Fe2O3_Bands.PlotShow(ymin=-2,ymax=4)
```

```python
# NELECT = 68
print("Spin Up:",Fe2O3_Bands.getBandsGap(band1=34,band2=35,bandspin="up"))
print("Spin Down : ",Fe2O3_Bands.getBandsGap(band1=34,band2=35,bandspin="down"))
```

```python
# Atom = 1 or 3 are the iron atoms with magmom = + 4.165
# Atom = 2 or 4 are the iron atoms with magmom = - 4.165
# Atom = 5-10 are the oxygen atoms.

# Plotting the. Projected Band of s, p, d (up) orbitals of Fe atom (atom=1).
s = 200
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_up",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_up",atom=1,scale=s,alpha=0.5,color="r",label="Fe($p_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_up",atom=1,scale=s,alpha=0.5,color="b",label="Fe($s_{up}$)")
Fe2O3_Bands.PlotShow()

#reduce y-range
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_up",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_up",atom=1,scale=s,alpha=0.5,color="r",label="Fe($p_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_up",atom=1,scale=s,alpha=0.5,color="b",label="Fe($s_{up}$)")
Fe2O3_Bands.PlotShow(ymin=-5,ymax=3)
```

```python
# Plotting the. Projected Band of s, p, d (down) orbitals of Fe atom (atom=1).
s = 200
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_down",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_down",atom=1,scale=s,alpha=0.5,color="r",label="Fe($p_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_down",atom=1,scale=s,alpha=0.5,color="b",label="Fe($s_{down}$)")
Fe2O3_Bands.PlotShow()

#reduce y-range
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_down",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_down",atom=1,scale=s,alpha=0.5,color="r",label="Fe($p_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_down",atom=1,scale=s,alpha=0.5,color="b",label="Fe($s_{down}$)")
Fe2O3_Bands.PlotShow(ymin=-5,ymax=3)
```

```python
# Plotting the. Projected Band of s, p, d  orbitals of Fe atom (atom=1) and O atom (atom=8).

# Spin Up
s = 500
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_up",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_up",atom=8,scale=s,alpha=0.5,color="b",label="O($p_{up}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_up",atom=8,scale=s,alpha=0.5,color="r",label="O($s_{up}$)")
Fe2O3_Bands.PlotShow(ymin=-3,ymax=3)

# Spin down
s = 500
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getOrbitalBandsPlot(orbital="d_down",atom=1,scale=s,alpha=0.5,color="g",label="Fe($d_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="p_down",atom=8,scale=s,alpha=0.5,color="b",label="O($p_{down}$)")
Fe2O3_Bands.getOrbitalBandsPlot(orbital="s_down",atom=8,scale=s,alpha=0.5,color="r",label="O($s_{down}$)")
Fe2O3_Bands.PlotShow(ymin=-3,ymax=3)
```

```python
# Plotting the. Total Projected Bands of Fe atoms (atom=1,2,3,4) and O atoms (atom=5,6,7,8,9,10).

# Spin Up
s = 10
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getAtomsRangeBandsPlot(AtomRange=[1,2,3,4],orbital="tot_up",scale=s,color="r",alpha=0.5,label="$Fe_{tot}(up)$")
Fe2O3_Bands.getAtomsRangeBandsPlot(AtomRange=[5,6,7,8,9,10],orbital="tot_up",scale=s,color="b",alpha=0.5,label="$O_{tot}(up)$")
Fe2O3_Bands.PlotShow(ymin=-3,ymax=3)


# Spin Down
# for AtomRange, one can be use the range function range(1,5) = [1,2,3,4]
s =10
Fe2O3_Bands.Init_Fig(width=5,height=10)
Fe2O3_Bands.getAtomsRangeBandsPlot(AtomRange=range(1,5),orbital="tot_down",scale=s,color="r",alpha=0.5,label="$Fe_{tot}(down)$")
Fe2O3_Bands.getAtomsRangeBandsPlot(AtomRange=range(5,11),orbital="tot_down",scale=s,color="b",alpha=0.5,label="$O_{tot}(down)$")
Fe2O3_Bands.PlotShow(ymin=-3,ymax=3)
```
